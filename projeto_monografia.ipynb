```python
import sidrapy
import pandas as pd
```

```python
data = sidrapy.get_table(table_code="2493", territorial_level="3", ibge_territorial_code="all", period="last", classifications={'2': '4,5', '12262': '104896,104897,104898,104899', '58': '104893,104894,104895,6798,3245'})
data = data.rename(columns=data.iloc[0]).iloc[1:].reset_index(drop=True)
data = data.iloc[:, [i for i in range(data.shape[1])
                          if (i < 4 and i % 2 != 0)
                          or (i >= 4 and i % 2 == 0)
                         ]]
data['Valor'].replace('-', '0', inplace=True)
```

```python
display(data)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Change column type to float32 for column: 'Valor'
    data = data.astype({'Valor': 'float32'})
    # Performed 1 aggregation grouped on column: 'Cobertura de plano de saúde'
    data = data.groupby(['Cobertura de plano de saúde']).agg(Valor_sum=('Valor', 'sum')).reset_index()
    return data

data_clean = clean_data(data.copy())
data_clean.head()
```
